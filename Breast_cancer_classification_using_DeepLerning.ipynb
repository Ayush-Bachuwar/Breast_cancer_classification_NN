```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib as plt
import sklearn.datasets
import matplotlib.pyplot as plt
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data,columns = breast_cancer_dataset.feature_names)
```

```python
data_frame['target'] = breast_cancer_dataset.target
```

Separating features and target

```python
x = data_frame.drop(columns = 'target', axis = 1)
y = data_frame['target']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)  #based on the scaling of train data test data will be standardize 
```

Building neural network

```python
import tensorflow as tf
tf.random.set_seed(3)    #to get same accuracy score
from tensorflow import keras
```

model

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)), #converting the data into 1D
    keras.layers.Dense(32, activation = 'relu'),
    keras.layers.Dense(2, activation = 'sigmoid')#1 unit for each class
])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = 'accuracy')
```

```python
history = model.fit(x_train_std,y_train, validation_split = 0.1, epochs = 10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy' ])
plt.title('model_accuracy')
plt.ylabel("accuracy")
plt.legend(["training data, validation data"], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss' ])
plt.title('model_accuracy')
plt.ylabel("accuracy")
plt.legend(["training data, validation data"], loc = 'lower right')
```

Accuracy of the model on testing data

```python
loss,accuracy  = model.evaluate(x_test_std,y_test)
print(loss, accuracy)
```

```python
y_pred = model.predict(x_test_std)
```
