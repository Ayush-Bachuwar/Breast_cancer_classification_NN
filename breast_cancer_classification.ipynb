```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

Data colletion and Data pre-processing

```python
dataset = pd.read_csv("data.csv")
dataset
```

```python
dataset = dataset.drop("Unnamed: 32",axis = 1)
```

```python
target = dataset.pop("diagnosis")
dataset.insert(31,"diagnosis",target)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset['diagnosis'].value_counts()
```

B --> Benign (safe not to worry)
M --> Malignant (A tumor which spreads to other tissues or organs DANGEROUS

```python
dataset.groupby('diagnosis').mean()
```

```python
x = dataset.drop(['diagnosis','id'], axis = 1) #removing id because it is of no use for training a model and diagnosis is the target we want to predict
y = dataset['diagnosis']
```

Splitting the data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2 )
```

```python
print(x_train.shape,x_test.shape)
```

Model training

```python
model = LogisticRegression()
model1 = DecisionTreeClassifier()
model.fit(x_train,y_train)
model1.fit(x_train,y_train)
```

Model Evaluation

```python
prediction = model.predict(x_train)
score = accuracy_score(y_train,prediction)
print("Accuracy using Logistic regression on the training data:",score) 
prediction1 = model1.predict(x_train)
score1 = accuracy_score(y_train,prediction1)
print("Accuracy using Decision tree classifier the training data:",score1) 
```

```python
prediction = model.predict(x_test)
score = accuracy_score(y_test,prediction)
print("Accuracy using Logistic regression on the training data:",score) 
prediction1 = model1.predict(x_test)
score1 = accuracy_score(y_test,prediction1)
print("Accuracy using Decision tree classifier the training data:",score1) 
```
